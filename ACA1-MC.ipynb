#### 1. Criteria
According to Alex, the school must be safe, urban, diverse and high-quality. I set some criteria for finding the schools that meets her requirements.
1.	Safety: The most common confirmation of safety in a city is the low crime rate, so I will calculate the data (sum of cases/population) from FBI's latest crime report, then exclude region of high crime rate (crime rate>0.1).
2.	Urban: Alex wants to live the big city life, so I will select the schools only located in a large city, and the school also should not be distance education only.
3.	Diversity: Alex wants to study in a diverse culture, which means the school must have people (both students and faculties) with different culture backgrounds. It means that we should check the student body by race. Those schools that have high preference (the enrollment percentage of a specific race >0.75) will be excluded from the list.
4.	Quality: First of all, Alex wants to pursue a degree in Education, so the schools must have the degrees awarded in Education (the percentage must not be 0). Second, the completion rate of the school is important, which reflects if the program is well-designed for students, as well as well-taught by teachers. Third, the students’ retention rate of the school is associated with students’ experience at the institution, which can also explain the quality of the program. I will select the schools which can award the degree in Education, and then look for the high-quality school with high completion rate (>0.7) and high retention rate (>0.7).   
5.Last but not least, the school must be currently operating.

#### 2.1 college scorecard dataset

```python
# BasicInfo: 'INSTNM','INSTURL','CITY','SARBBR',CURROPER' Columns for basic information of the shcool, which is useful for further investigation
# Urban: 'LOCALE','DISTANCEONLY' Columns for the criteria of urban
# Diversity: 'UGDS_WHITE','UGDS_BLACK','UGDS_HISP','UGDS_ASIAN','UGDS_AIAN','UGDS_NHPI','UGDS_2MOR','UGDS_NRA' Columns for the criteria of diversity
# Quality: 'C150_4','C150_L4','RET_FT4','RET_FTL4','PCIP13' Columns for the criteria of quality
subsetcols=['INSTNM','INSTURL','CITY','STABBR','CURROPER','LOCALE','DISTANCEONLY','UGDS_WHITE','UGDS_BLACK','UGDS_HISP','UGDS_ASIAN','UGDS_AIAN','UGDS_NHPI','UGDS_2MOR','UGDS_NRA','C150_4','C150_L4','RET_FT4','RET_FTL4','PCIP13']
CSCdata=pd.read_csv("CollegeScorecard.csv",usecols=subsetcols)
# only import columns that are useful for analysis
CSCdata
```

#### 2.2 crime dataset

```python
# Befor loading the dataset, I did some changes in Excel:
# 1.delete the useless rows (like titles of the dataset, and some explanations for the data)
# 2.set the numerical type for some columns
# 3.do some comlementary work for the first column of State. For better merging the two dataset: replace the state name with their abbreiations, and then drag it to fill the blank.
# 4.For better merging with the CSCdataset, deleting the nubers appear in the column 2 (CITY, which is used as annotation) by using the replace function in Excel.
```

```python
Crimedata=pd.read_csv("Crime2019.csv")
Crimedata=Crimedata.fillna(0) # Fill the blank in the list (Supposed that the blank represents 0)
Crimedata
```

#### 3 Refine and merge data

#### 3.1 CSCdata

```python
# First of all, only schools that are currently operating should be selected.
CSC=CSCdata[CSCdata['CURROPER']==1]
```

```python
# For the criteria of urban, only schools that are in large cities and not distance only can be selected.
CSC=CSC[CSC['LOCALE']==11]
CSC=CSC[CSC['DISTANCEONLY']==0]
CSC
```

```python
# For the criteria of quality, the schools should be able to award the degree in Education
CSC=CSC[CSC['PCIP13']>0]
# the schools should also have high completion rate and retention rate (>0.7) (the lengths for the program is not associated with the quality, so both four-year schools and less-than-four year schools are counted)
CSC=CSC[(CSC['C150_4']>0.7)|(CSC['C150_L4']>0.7)]
CSC=CSC[(CSC['RET_FT4']>0.7)|(CSC['RET_FTL4']>0.7)]
```

```python
# For the criteria of cultural diversity, drop the schools with high preference of a specific race
CSC=CSC[(CSC['UGDS_WHITE']<0.75)&(CSC['UGDS_BLACK']<0.75)&(CSC['UGDS_HISP']<0.75)&(CSC['UGDS_ASIAN']<0.75)&(CSC['UGDS_AIAN']<0.75)&(CSC['UGDS_NHPI']<0.75)&(CSC['UGDS_2MOR']<0.75)&(CSC['UGDS_NRA']<0.75)]

# Create a new column for merging 'CITY' and 'STABBR', which helps locate the specific city.
CSC['city']=CSC['CITY']+[',']+CSC['STABBR']

## Create new columns to refine information for further investigation and ranking
# Create a new column to represent school quality (aggregate the completion rate and retention rate)
CSC=CSC.fillna(0)
CSC['quality']=CSC['C150_4']+CSC['C150_L4']+CSC['RET_FT4']+CSC['RET_FTL4']
# Create a new column to represent diversity (the lower the percentage of the largest race group of the school, the more diverse the school should be)
CSCslice=CSC[['UGDS_WHITE','UGDS_BLACK','UGDS_HISP','UGDS_ASIAN','UGDS_AIAN','UGDS_NHPI','UGDS_2MOR','UGDS_NRA']]
CSC['diversity']=1-CSCslice.max(axis=1)

# Refine the dataset
CSC=CSC[['INSTNM','INSTURL','city','quality','diversity']]
CSC
```

#### 3.2 Crimedata

```python
Crimedata
```

```python
# Calculate the crime rate of each city, and create a new column for crime rate
Crimedata['CrimeRate']=(Crimedata['Violent']+Crimedata['manslaughter']+Crimedata['Rape1']+Crimedata['Robbery']+Crimedata['assault']+Crimedata['Property']+Crimedata['Burglary']+Crimedata['LarcenyTheft']+Crimedata['MotorVehicleTheft']+Crimedata['Arson2'])/Crimedata['Population']

# Create a new column for locating the city
Crimedata['city']=Crimedata['City']+[',']+Crimedata['State']

Crime=Crimedata[['city','CrimeRate']]

Crime
```

#### 3.3 merge data

```python
Result=pd.merge(CSC,Crime)
# For the criteria of safety, exclude the schoold with crimerate >0.1
Result=Result[Result['CrimeRate']<0.1]
Result
```

#### 4 Rank

```python
# The rank of the school should consider all criteria, so I will rank the quality, diversity, and crimerate separately, then use the sum of the different ranks to show the final rank
Result['rankQ']=Result.quality.rank(method='dense',ascending=False)
Result['rankD']=Result.diversity.rank(method='dense',ascending=False)
Result['rankS']=Result.CrimeRate.rank(method='dense',ascending=True)
Result['Sumrank']=Result['rankQ']+Result['rankD']+Result['rankS']
Result['Rank']=Result.Sumrank.rank(method='dense',ascending=True)
# Sort the list according to the rank
Result=Result.sort_values(by='Rank',ascending=True)
Result=Result[['INSTNM','INSTURL','city','quality','diversity','CrimeRate','Rank']]
Result
```

#### 5 Output dataset

```python
Result.to_csv('Result.csv',index=False)
# set index as false to delete the first column (the mixed numbers)
```
